```python
import pandas as pd 
value = { "Name":["Atharva","Dipti","Omkar","Narayan"],"Age":[22,21,24,19], "Salary":[10000,90000,1912,12721],"Department":["IT","CS","ENTC","IT"]}
df = pd.DataFrame(value)
df
```

```python
#All name from the DataFrame 
print(df["Name"])
```

```python
# Question:
# You have a DataFrame containing employee details with columns:
# Name, Department, Salary, Age.

# Write code to:

# Display employees whose salary is greater than 100,00.
print("Display employees whose salary is greater than 100,00.\n",df[df["Salary"]>10000])
# Display employees from the IT department.
print("Display employees from the IT department.\n",df[df["Department"]=="IT"])
# Sort the employees by salary in descending order.
print("Sort the employees by salary in descending order.\n",df["Salary"].sort_values(ascending=False).reset_index(drop=True))

```
