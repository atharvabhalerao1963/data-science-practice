```python
# 8 july 
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [23, 30, 22, 28, 35],
    "Salary": [50000, 70000, 45000, 65000, 80000]
}

df = pd.DataFrame(data)
# Tasks
# Display employees whose age is greater than 25.
# Display employees whose salary is at least 65,000.
# Display only the Name and Salary columns for employees older than 25.

age = df[df["Age"]>25]
salary = df[df["Salary"]>=65000]
result = df.loc[df["Age"] > 25, ["Name", "Salary"]]
print("Display employees whose age is greater than 25.",age)
print("Display employees whose salary is at least 65,000.",salary)
print("Display only the Name and Salary columns for employees older than 25.",result)
```

```python
# 9 july  
#  Find the Highest Paid Employee in Each Department
import pandas as pd

data = {
    "Name": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Sneha"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "Salary": [50000, 45000, 70000, 60000, 55000, 75000]
}

df = pd.DataFrame(data)
# Interview Question

# Find the employee with the highest salary in each department.
df.groupby("Department")[["Salary","Name"]].max()

```

```python
# : Add a New Column
# import pandas as pd

data = {
    "Name": ["Amit", "Neha", "Rahul", "Priya"],
    "Salary": [40000, 60000, 55000, 80000]
}

df = pd.DataFrame(data)

# Create a new column called Bonus.

# If salary is greater than or equal to 60000 → Bonus = 10000
# Otherwise → Bonus = 5000
df["Bonus"] = df["Salary"].apply(lambda x : 10000 if x >= 60000 else 5000)
print(df)
```
